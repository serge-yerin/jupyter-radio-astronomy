```python
from matplotlib import pyplot as plt
import numpy as np

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

x_values = np.linspace(-3, 3, 1200)
mu = 0
sig = 1
y_values = gaussian(x_values, mu, sig)

plt.figure(1, figsize=(12.0, 8.0))
plt.plot(x_values, y_values)
plt.hlines(0.9, -3, 3,   colors='C1')
plt.hlines(0.7, -3, 3,   colors='C2' )
plt.hlines(0.5, -3, 3,   colors='C3')
plt.vlines(0,    0, 1.1, colors='C0', linestyle=':')

order = y_values.argsort()
y_values = y_values[order]
x_values = x_values[order]

def what_is_x_when_y_is(input, x, y):
    return abs(x[y.searchsorted(input, 'left')])

def interp_x_from_y(input, x, y):
    return np.interp(input, y, x)

plt.vlines( what_is_x_when_y_is(0.5, x_values, y_values),    0, 1.1, colors='C3')
plt.vlines( what_is_x_when_y_is(0.7, x_values, y_values),    0, 1.1, colors='C2')
plt.vlines( what_is_x_when_y_is(0.9, x_values, y_values),    0, 1.1, colors='C1')
plt.vlines(-what_is_x_when_y_is(0.5, x_values, y_values),    0, 1.1, colors='C3')
plt.vlines(-what_is_x_when_y_is(0.7, x_values, y_values),    0, 1.1, colors='C2')
plt.vlines(-what_is_x_when_y_is(0.9, x_values, y_values),    0, 1.1, colors='C1')

plt.show()
```

```python
ratio9_7 = what_is_x_when_y_is(0.9, x_values, y_values) / what_is_x_when_y_is(0.7, x_values, y_values)
ratio9_5 = what_is_x_when_y_is(0.9, x_values, y_values) / what_is_x_when_y_is(0.5, x_values, y_values)
ratio7_5 = what_is_x_when_y_is(0.7, x_values, y_values) / what_is_x_when_y_is(0.5, x_values, y_values)

print('\n')
print('  Length of 0.9 level on x scale is: ', what_is_x_when_y_is(0.9, x_values, y_values))
print('  Length of 0.7 level on x scale is: ', what_is_x_when_y_is(0.7, x_values, y_values))
print('  Length of 0.5 level on x scale is: ', what_is_x_when_y_is(0.5, x_values, y_values))

print('\n')
print('  Ratio of 0.9 and 0.7 levels on x scale is: ', ratio9_7)
print('  Ratio of 0.9 and 0.5 levels on x scale is: ', ratio9_5)
print('  Ratio of 0.7 and 0.5 levels on x scale is: ', ratio7_5)
```

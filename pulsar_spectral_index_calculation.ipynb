# Calculation of spectral index based on two frequency points

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1 = 25   # MHz
s1 = 3    # mJy
f2 = 120  # Mhz
s2 = 8    # mJy
```



Common formula for calculation of spectral index based on two frequency points:  <font size="5">  $  \alpha = \frac{\log_{10}{\frac{s_2}{s_1}}}{\log_{10}{\frac{f_2}{f_1}}}  $ </font>



```python
sp_index = ((np.log10(s2/s1))/(np.log10(f2/f1)))
print('\n\n Spectral index = ', sp_index, '\n\n')
```

```python
f = np.linspace(10, 130, 120)
model_flux = np.power(f, sp_index)/2.5       #22000
```

```python
plt.figure(figsize = (18, 6))

plt.subplot(121)
plt.title('Linear scales on axis')
plt.plot ([f1, f2],[s1, s2], 'ro')
plt.plot (f, model_flux)

plt.subplot(122)
plt.title('Logarythmic scales on axis')
plt.plot ([f1, f2],[s1, s2], 'ro')
plt.plot (f, model_flux)
plt.xscale('log')
plt.yscale('log')
plt.show()
```

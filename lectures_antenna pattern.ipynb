# Antenna pattern demonstration



### Import necessary liraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
angles = np.linspace(-180, 180, 362)
```

```python
phi = 1
pattern_1 = np.abs(np.sin(0.2 * angles + phi) / (0.2 * angles + phi))
phi = 2
pattern_2 = np.abs(np.sin(0.2 * angles + phi) / (0.2 * angles + phi))
phi = 3
pattern_3 = np.abs(np.sin(0.2 * angles + phi) / (0.2 * angles + phi))
phi = 4
pattern_4 = np.abs(np.sin(0.2 * angles + phi) / (0.2 * angles + phi))
phi = 5
pattern_5 = np.abs(np.sin(0.2 * angles + phi) / (0.2 * angles + phi))
pattern_1 = pattern_1 / np.max(pattern_1)
pattern_2 = pattern_2 / np.max(pattern_2)
pattern_3 = pattern_3 / np.max(pattern_3)
pattern_4 = pattern_4 / np.max(pattern_4)
pattern_5 = pattern_5 / np.max(pattern_5)
```

```python
plt.figure(figsize=(15, 7))
plt.plot(angles, pattern_1)
plt.plot(angles, pattern_2)
plt.plot(angles, pattern_3)
plt.plot(angles, pattern_4)
plt.plot(angles, pattern_5)
plt.show()
```

```python
theta = (angles + 90) * np.pi / 180

plt.figure(figsize=(15, 15))
ax = plt.subplot(111, projection='polar')
ax.plot(theta, pattern_1)
ax.plot(theta, pattern_2)
ax.plot(theta, pattern_3, linewidth=5)
ax.plot(theta, pattern_4)
ax.plot(theta, pattern_5)
ax.set_rmax(1)
ax.set_thetamin(0)
ax.set_thetamax(180)
#ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(22.5)  # Move radial labels away from plotted line
ax.grid(True)
#ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```

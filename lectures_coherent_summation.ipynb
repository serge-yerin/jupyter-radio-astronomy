```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Npoints = 1000
tmin = 0
tmax = 30
dt = 0.01
Fs = 1 / dt

points = np.linspace(tmin, tmax, Npoints)
spectrum_length = Npoints // 2
```

```python
harmonic_01 = 2.0 * np.sin(2.0 * points)
harmonic_02 = 1.6 * np.sin(2.0 * points)
coherent_sum = harmonic_01 + harmonic_02
```

```python
fig = plt.figure(facecolor = 'white', figsize = (16, 10))
plt.subplot(311)
plt.plot(points, harmonic_01, linewidth = 5, color = 'C0')
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
plt.subplot(312)
plt.plot(points, harmonic_02, linewidth = 5, color = 'C1')
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
plt.subplot(313)
plt.plot(points, coherent_sum, linewidth = 5, color = 'C3')
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
```

```python
power_01 = harmonic_01 ** 2
power_02 = harmonic_02 ** 2
power_incoh_sum = power_01 + power_02
power_coh_sum = coherent_sum ** 2
```

```python
fig = plt.figure(facecolor = 'white', figsize = (16, 6))
plt.subplot(111)
plt.plot(points, power_01, linewidth = 3, color = 'C0')
plt.plot(points, power_02, linewidth = 3, color = 'C1')
plt.plot(points, power_coh_sum, linewidth = 3, color = 'C3')
plt.axis([tmin, tmax, -1, 15])
plt.grid(True)
```

```python
fig = plt.figure(facecolor = 'white', figsize = (16, 6))
plt.subplot(111)
plt.plot(points, power_01, linewidth = 3, color = 'C0')
plt.plot(points, power_02, linewidth = 3, color = 'C1')
plt.plot(points, power_coh_sum, linewidth = 3, color = 'C3')
plt.plot(points, power_incoh_sum, linewidth = 3, color = 'C2')
plt.axis([tmin, tmax, -1, 15])
plt.grid(True)
```

```python
fig = plt.figure(facecolor = 'white', figsize = (8, 6))
plt.subplot(111)
plt.plot(points, np.ones_like(points) * np.mean(power_01), linewidth = 5, color = 'C0')
plt.plot(points, np.ones_like(points) * np.mean(power_02), linewidth = 5, color = 'C1')
plt.plot(points, np.ones_like(points) * np.mean(power_coh_sum), linewidth = 5, color = 'C3')
plt.plot(points, np.ones_like(points) * np.mean(power_incoh_sum), linewidth = 5, color = 'C2')
plt.axis([tmin, tmax, -1, 10])
plt.grid(True)
```

```python
q = psrqpy.QueryATNF()
numstring = 'Version {} of the ATNF catalogue contains {} pulsars'
print(numstring.format(q.get_version, q.num_pulsars))
full_catalogue = q.catalogue
```

```python
print(full_catalogue)
```

```python
for i in range(10):
    print("{0:15s}   {1:24.18f} s".format(full_catalogue["PSRJ"][i], float(full_catalogue["P0"][i])))
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(18,6)) 
ax0.hist(full_catalogue["P0"], bins=100, alpha=0.9, log=True);
ax0.set_xlabel('Period, s')
ax0.set_ylabel('Number of pulsars')
ax1.hist(full_catalogue["F0"], bins=100, alpha=0.9, log=True);
ax1.set_xlabel('Frequency, Hz')
ax1.set_ylabel('Number of pulsars')
```

```python
# Best estimate of the pulsar distance using the YMW16 DM-based distance as default (kpc)
for i in range(10):
    print("{0:15s}   {1:8.4f} kpc".format(full_catalogue["PSRJ"][i], float(full_catalogue["DIST"][i])))
```

```python
fig, ax0 = plt.subplots(1, 1, figsize=(18,6)) 
ax0.hist(full_catalogue["DIST"] * 3261.5637769443, bins=500, alpha=0.9);  # convert kpc to ly
ax0.axvline(x = 52850, color='red', alpha=0.9, label = 'Milky Way radius');
ax0.set_xlabel('Distance, light years')
ax0.set_ylabel('Number of pulsars')
ax0.legend()
```

```python
print("\n The nearest  pulsar {0:14s} is located {1:8.4f} kpc    or {2:12.4f} ly away from Earth, DM = {3:10.5f}".format(full_catalogue.loc[full_catalogue["DIST"].idxmin()]['PSRJ'], 
                                                                                                       full_catalogue["DIST"].min(),
                                                                                                       full_catalogue["DIST"].min() * 3261.5637769443,
                                                                                                       full_catalogue.loc[full_catalogue["DIST"].idxmin()]['DM']))

print(" The farthest pulsar {0:14s} is located {1:8.4f} kpc    or {2:12.4f} ly away from Earth, DM = {3:10.5f}".format(full_catalogue.loc[full_catalogue["DIST"].idxmax()]['PSRJ'], 
                                                                                                       full_catalogue["DIST"].max(),
                                                                                                       full_catalogue["DIST"].max() * 3261.5637769443,
                                                                                                       full_catalogue.loc[full_catalogue["DIST"].idxmax()]['DM']))

print("\n The min DM = {0:10.4f} pulsar is {1:14s} located {2:8.4f} kpc    or {3:12.4f} ly away from Earth".format(full_catalogue["DM"].min(),
                                                                                                                 full_catalogue.loc[full_catalogue["DM"].idxmin()]['PSRJ'],
                                                                                                                 full_catalogue.loc[full_catalogue["DM"].idxmin()]['DIST'],
                                                                                                                 full_catalogue.loc[full_catalogue["DM"].idxmin()]['DIST'] * 3261.5637769443,
                                                                                                                ))

print(" The max DM = {0:10.4f} pulsar is {1:14s} located {2:8.4f} kpc    or {3:12.4f} ly away from Earth".format(full_catalogue["DM"].max(),
                                                                                                                 full_catalogue.loc[full_catalogue["DM"].idxmax()]['PSRJ'],
                                                                                                                 full_catalogue.loc[full_catalogue["DM"].idxmax()]['DIST'],
                                                                                                                 full_catalogue.loc[full_catalogue["DM"].idxmax()]['DIST'] * 3261.5637769443,
                                                                                                                ))




```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
c = 299792458                    # Speed of light, m/s
kB = 1.38064852 * pow(10,-23)    # Boltzmann constant, W * m-2 * Hz-1 * sr-1

freq = np.linspace(8, 80, 73)
```

```python
Igal = 2.48*pow(10,-20)*pow(freq,-0.52)*(1-np.exp(-5*pow(freq,-2.1)))/(5*pow(freq,-2.1))
Iexgal = 1.06*pow(10,-20)*pow(freq,-0.8)*np.exp(-5*pow(freq,-2.1))
Tbg_Cane = (Iexgal + Igal) * pow(c, 2) / (2 * pow(freq*1000000, 2) * kB)
Tbg_Krymkin = 3.78*pow(10,5) * pow((10/freq), 2.56)
```

```python
BG_temp = pd.DataFrame({
    'Frequency':      [  10.0,   12.6,   14.7,   16.7,   20.0,  25.0, 45.0  ],
    'Tbg_B0031-07':   [np.nan, 137000,  62000,  48000,  26000, 19000, np.nan],
    'Tbg_В1133+16':   [np.nan, np.nan, np.nan, np.nan, np.nan, 24000, 8000  ],
    'Tbg_В1508+55':   [np.nan, np.nan, np.nan, np.nan, np.nan, 22000, 8000  ],
    'Tbg_В1919+21':   [np.nan, np.nan, np.nan, np.nan, np.nan, 42000, 12000 ],
    'Tbg_J0250+5854': [353000, 174000, 134000, 110000,  67000, 41000, 11000 ]})
```

```python
plt.figure(facecolor = 'white', figsize = (16, 10))
ax = plt.subplot(111)

plt.plot(freq, Tbg_Krymkin, color = 'C3', label = 'Krymkin model', lw = 2)
plt.plot(freq, Tbg_Cane, color = 'C0', label = 'Cane model', lw = 4, linestyle = ':')

plt.scatter(BG_temp['Frequency'], BG_temp['Tbg_B0031-07'],   color = 'C0', label = 'Tbg B0031-07',   marker='o', lw = 4)
plt.scatter(BG_temp['Frequency'], BG_temp['Tbg_В1133+16'],   color = 'C1', label = 'Tbg В1133+16',   marker='o', lw = 4)
plt.scatter(BG_temp['Frequency'], BG_temp['Tbg_В1508+55'],   color = 'C2', label = 'Tbg В1508+55',   marker='o', lw = 4)
plt.scatter(BG_temp['Frequency'], BG_temp['Tbg_В1919+21'],   color = 'C3', label = 'Tbg В1919+21',   marker='o', lw = 4)
plt.scatter(BG_temp['Frequency'], BG_temp['Tbg_J0250+5854'], color = 'C4', label = 'Tbg J0250+5854', marker='o', lw = 4)

plt.legend(loc = 'upper right', fontsize = 18)
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(3.0)
ax.xaxis.set_tick_params(width=3, which='both', direction='in', length=10, labelsize = 20)
ax.yaxis.set_tick_params(width=3, which='major', direction='in', length=10, labelsize = 20)
ax.yaxis.set_tick_params(width=3, which='minor', direction='in', length=7, labelsize = 20)
plt.xlabel('Frequency, MHz', fontsize=25)
plt.ylabel('Brightness temperature, K', fontsize=25)
plt.axis([7, 85, 1000,  400000])

```

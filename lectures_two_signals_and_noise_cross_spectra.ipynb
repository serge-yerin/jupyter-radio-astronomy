```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Npoints = 1000
tmin = 0
tmax = 30
dt = 0.01
Fs = 1 / dt

points = np.linspace(tmin, tmax, Npoints)
spectrum_length = Npoints // 2
```

```python
harmonic_01 = 3.0 * np.sin(1.0 * points)
harmonic_02 = 0.8 * np.sin(8.0 * points)
harmonic_03 = 2.5 * np.sin(2.5 * points)
signal_mix_01 = harmonic_01 + harmonic_02
signal_mix_02 = harmonic_02 + harmonic_03
```

```python
fig = plt.figure(facecolor = 'white', figsize = (16, 5))
plt.subplot(131)
plt.plot(points, harmonic_01)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
plt.subplot(132)
plt.plot(points, harmonic_02)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
plt.subplot(133)
plt.plot(points, harmonic_03)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
```

```python
fig = plt.figure(facecolor = 'white', figsize = (16, 5))
plt.subplot(121)
plt.plot(points, signal_mix_01)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
plt.subplot(122)
plt.plot(points, signal_mix_02)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
```

```python
spectra_01 = np.fft.fft(harmonic_01)[0:spectrum_length]

plt.figure(1, figsize = [16, 6], facecolor = 'white')
plt.subplot(121)
plt.plot(np.abs(spectra_01))
# plt.axis([-2, 52, -0.1, 1.65])
plt.grid(True)
plt.subplot(122)
plt.plot(np.abs(spectra_01))
# plt.axis([-2, 52, -150, 10])
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
spectra_mix_1 = np.fft.fft(signal_mix_01)[0:spectrum_length]
spectra_mix_2 = np.fft.fft(signal_mix_02)[0:spectrum_length]

plt.figure(1, figsize = [16, 8], facecolor = 'white')
plt.subplot(221)
plt.plot(np.abs(spectra_mix_1))
# plt.axis([-2, 52, -0.1, 1.65])
plt.grid(True)
plt.subplot(222)
plt.plot(np.abs(spectra_mix_1))
# plt.axis([-2, 52, -150, 10])
plt.yscale('log')
plt.grid(True)
plt.subplot(223)
plt.plot(np.abs(spectra_mix_2))
# plt.axis([-2, 52, -0.1, 1.65])
plt.grid(True)
plt.subplot(224)
plt.plot(np.abs(spectra_mix_2))
# plt.axis([-2, 52, -150, 10])
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
mean = 0
std = 1 
noise_01 = np.random.normal(mean, std, len(points))
noise_02 = np.random.normal(mean, std, len(points))
noise_03 = np.random.normal(mean, std, len(points))
noise_04 = np.random.normal(mean, std, len(points))
noise_spectrum_01 = np.fft.fft(noise_01)[0:spectrum_length]
noise_spectrum_02 = np.fft.fft(noise_02)[0:spectrum_length]
noise_spectrum_03 = np.fft.fft(noise_03)[0:spectrum_length]
noise_spectrum_04 = np.fft.fft(noise_04)[0:spectrum_length]
```

```python
fig = plt.figure(facecolor = 'white', figsize = (16, 5))
plt.subplot(121)
plt.plot(points, noise_01)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
plt.subplot(122)
plt.plot(np.abs(noise_spectrum_01))
#plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
```

```python
noise_mix_01 = signal_mix_01 + 1.0 * noise_01
noise_mix_02 = signal_mix_01 + 5.0 * noise_02
noise_mix_03 = signal_mix_02 + 1.0 * noise_03
noise_mix_04 = signal_mix_02 + 5.0 * noise_04
```

```python
fig = plt.figure(facecolor = 'white', figsize = (16, 10))
plt.subplot(221)
plt.plot(points, noise_mix_01)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
plt.subplot(222)
plt.plot(points, noise_mix_02)
plt.axis([tmin, tmax, -15, 15])
plt.grid(True)
plt.subplot(223)
plt.plot(points, noise_mix_03)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
plt.subplot(224)
plt.plot(points, noise_mix_04)
plt.axis([tmin, tmax, -15, 15])
plt.grid(True)
```

```python
spectra_nmix_1 = np.fft.fft(noise_mix_01)[0:spectrum_length]
spectra_nmix_2 = np.fft.fft(noise_mix_02)[0:spectrum_length]
spectra_nmix_3 = np.fft.fft(noise_mix_03)[0:spectrum_length]
spectra_nmix_4 = np.fft.fft(noise_mix_04)[0:spectrum_length]

plt.figure(1, figsize = [16, 12], facecolor = 'white')
plt.subplot(421)
plt.plot(np.abs(spectra_nmix_1))
# plt.axis([-2, 52, -0.1, 1.65])
plt.grid(True)
plt.subplot(422)
plt.plot(np.abs(spectra_nmix_1))
# plt.axis([-2, 52, -150, 10])
plt.yscale('log')
plt.grid(True)

plt.subplot(423)
plt.plot(np.abs(spectra_nmix_2))
# plt.axis([-2, 52, -0.1, 1.65])
plt.grid(True)
plt.subplot(424)
plt.plot(np.abs(spectra_nmix_2))
# plt.axis([-2, 52, -150, 10])
plt.yscale('log')
plt.grid(True)

plt.subplot(425)
plt.plot(np.abs(spectra_nmix_3))
# plt.axis([-2, 52, -0.1, 1.65])
plt.grid(True)
plt.subplot(426)
plt.plot(np.abs(spectra_nmix_3))
# plt.axis([-2, 52, -150, 10])
plt.yscale('log')
plt.grid(True)

plt.subplot(427)
plt.plot(np.abs(spectra_nmix_4))
# plt.axis([-2, 52, -0.1, 1.65])
plt.grid(True)
plt.subplot(428)
plt.plot(np.abs(spectra_nmix_4))
# plt.axis([-2, 52, -150, 10])
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
cross_spectra_2_3 = spectra_nmix_2 * np.conj(spectra_nmix_3)

plt.figure(1, figsize = [16,10], facecolor = 'white')
plt.subplot(221)
plt.plot(np.abs(cross_spectra_2_3))
# plt.axis([-2, 52, -0.1, 1.65])
plt.grid(True)
plt.subplot(222)
plt.plot(np.abs(cross_spectra_2_3))
# plt.axis([-2, 52, -150, 10])
plt.yscale('log')
plt.grid(True)
plt.show()
```

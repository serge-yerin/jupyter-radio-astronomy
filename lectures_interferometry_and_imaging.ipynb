```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal
```

```python
im = cv2.imread('DATA/sky.jpg')
im.shape
```

```python
plt.imshow(im[:,:,::-1])
plt.axis('off')
plt.show()
```

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im.shape
```

```python
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.show()
```

```python
print(np.min(im), np.max(im))
```

```python
im_sp = np.fft.fft2(im)
im_sp = np.fft.fftshift(im_sp)
im_sp = 10 * np.log(np.abs(im_sp))

# im_sp = im_sp - np.min(im_sp)
# im_sp = 255 * (im_sp / np.max(im_sp))

plt.imshow(im_sp, cmap='Greys_r')
plt.axis('off')
plt.show()
```

```python
print(np.min(im_sp), np.max(im_sp))
im_sp.shape
```

```python
im = cv2.imread('DATA/corr.png')
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

corr = signal.correlate2d(im, im, boundary='symm', mode='same')
plt.imshow(corr, cmap='Greys_r')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Npoints = 1000
tmin = 0
tmax = 30
dt = 0.01
Fs = 1/dt

points = np.linspace(tmin, tmax, Npoints)
```

```python
signal01 = 3*np.sin(points)
signal02 = 0.8 * np.sin(6*points)
signal = signal01 + signal02
```

```python
fig = plt.figure(facecolor = 'white', figsize = (12,4))
plt.subplot(121)
plt.plot(points, signal01)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
plt.subplot(122)
plt.plot(points, signal02)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
```

```python
fig = plt.figure(facecolor = 'white', figsize = (12, 4))
plt.subplot(121)
plt.plot(points, signal)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
```

```python
plt.figure(1, figsize = [12,7], facecolor = 'white')

plt.subplot(221)
plt.magnitude_spectrum(signal01, Fs = Fs)
plt.axis([-2, 52, -0.1, 1.65])
plt.grid(True)

plt.subplot(222)
plt.magnitude_spectrum(signal01, Fs=Fs, scale='dB')
plt.axis([-2, 52, -150, 10])
plt.grid(True)

plt.show()
```

```python
plt.figure(2, figsize = [12,7], facecolor = 'white')

plt.subplot(221)
plt.magnitude_spectrum(signal02, Fs=Fs)
plt.axis([-2, 52, -0.1, 1.65])
plt.grid(True)

plt.subplot(222)
plt.magnitude_spectrum(signal02, Fs=Fs, scale='dB')
plt.axis([-2, 52, -150, 10])
plt.grid(True)

plt.show()
```

```python
plt.figure(1, figsize = [12,7], facecolor = 'white')

plt.subplot(221)
plt.magnitude_spectrum(signal, Fs=Fs)
plt.axis([-2, 52, -0.1, 1.65])
plt.grid(True)

plt.subplot(222)
plt.magnitude_spectrum(signal, Fs=Fs, scale='dB')
plt.axis([-2, 52, -150, 10])
plt.grid(True)

plt.show()
```

```python
mean = 0
std = 1 
noise = np.random.normal(mean, std, len(points))
```

```python
fig = plt.figure(facecolor = 'white', figsize = (12, 4))
plt.subplot(121)
plt.plot(points, noise)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
```

```python
mix01 = signal + noise
mix02 = signal + 5.0*noise
```

```python
fig = plt.figure(facecolor = 'white', figsize = (12, 4))
plt.subplot(121)
plt.plot(points, mix01)
plt.axis([tmin, tmax, -5, 5])
plt.grid(True)
```
